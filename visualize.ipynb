```python
from joblib import dump, load
from help.utils import Timer
from help import helper as hp
from keras import backend as K
from keras.models import load_model
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import random
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
x = load('dataset/vectors/ori/all.jlb')[:,:-1]

autoencoder, encoder, model_str, _ = hp.buildDAE('2', x.shape[1], h_activation='sigmoid', o_activation='linear')
autoencoder.compile(optimizer='rmsprop', loss='mse')
autoencoder.summary()

timer = Timer().start()
history = autoencoder.fit(x=x, y=x, epochs=400, batch_size=256, verbose=1)
timer.end()

tstr = timer.summary(f=0)
encoder.save(f'model/dae/gpu.no generator.encoder.2.{tstr}.h5')
autoencoder.save(f'model/dae/gpu.no generator.autoencoder.2.{tstr}.h5')
dump(history, f'model/dae/gpu.no generator.history.2.{tstr}.jlb')

del x
del autoencoder
del encoder
K.clear_session()
```

```python
def visualisasi_data(x, y, judul='Visualisasi Data'):
    x = load_model('model/dae/gpu.no generator.encoder.2.19 menit 51 detik.h5').predict(x)
#     x = pca.transform(x)
    K.clear_session()
    
    colors = ['blue', 'green', 'red', 'brown', 'yellow', 'black', 'cyan', 'magenta']
    
    plt.figure(figsize=(11, 10))
    for c, label in enumerate(np.unique(y)):
        index = np.where(y == label)[0]
        plt.scatter(x=x[index,0], 
                    y=x[index,1], 
                    c=colors[c], 
                    label=label,
                    marker='o')
    plt.legend(bbox_to_anchor=(1,1))
    plt.title(judul)
    plt.show()
```

```python
expressions = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
expressions = np.array(expressions)
expressions
```

```python
name = 'ck+'
x = load(f'dataset/vectors/ori/{name}.jlb')
y = expressions[x[:,-1].astype(np.int)]
x = x[:,:-1]
visualisasi_data(x, y, judul=name)
del x
del y
```
